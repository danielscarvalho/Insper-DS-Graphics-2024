## Visualização

### Dados sobre veículos (cars dataset)

```python
import altair as alt
from vega_datasets import data

import hvplot.pandas 
import holoviews as hv
from holoviews import opts

cars_df = data.cars()
```

```python
cars_df.hvplot(
    by=['Horsepower'],
    kind='scatter',
    x='index',
    y=['Miles_per_Gallon'],
    legend='bottom_right',
    widget_location='bottom',
)
```

Gráfico PNG



```python
cars_df.hvplot.heatmap(x='Horsepower', 
                       y='Miles_per_Gallon', 
                       C='Acceleration', 
                       height=500, 
                       width=800, 
                       colorbar=True)
```
